```python
import itertools

def generate_passwords(min_length, max_length, output_file="wifi_password_wordlist.txt"):
    # Open the file to save the wordlist
    with open(output_file, "w") as file:
        for length in range(min_length, max_length + 1):
            # Generate all combinations of numbers for the current length
            for combo in itertools.product("0123456789", repeat=length):
                password = "".join(combo)
                file.write(password + "\n")
    print(f"Wordlist generated and saved to {output_file}")

# Customize length range for the passwords
min_length = 8  # Minimum length for passwords
max_length = 10  # Maximum length for passwords

generate_passwords(min_length, max_length)
```
